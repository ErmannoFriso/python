### Lookup tramite dataframe

```python
M01_CAS_post = pd.read_csv("test/after/M01_CAS.csv", sep="\t")
M01_CAS_post.drop(columns="Unnamed: 2", inplace=True)
M01_CAS_post["DataOra"] = pd.to_datetime(M01_CAS_post["DataOra"], format="%d/%m/%Y %H:%M:%S")
M01_CAS_post
```

```python
# Ricerca dei valori duplicati
M01_CAS_post.value_counts("Valore")
```

```python
#Inserimento in un Dataframe - da stringa non funziona
lookup = pd.DataFrame(M01_CAS_post.value_counts("Valore")).rename(columns={0: 'tentativi'})
lookup
```

```python
# Unione al dataframe originale per avere colonna con il numero di ripetizioni di ogni valore
M01_CAS_post2 = pd.merge(M01_CAS_post,lookup,on = "Valore",how = "inner")
M01_CAS_post2
```

## Scatter Plot with Seaborn
### Size + Hue per i tentativi

```python
sns.scatterplot(data=M01_CAS_post2[M01_CAS_post2["Valore"]!=0], x="DataOra", y="Valore", 
                size="tentativi", 
                hue="tentativi",
               palette="coolwarm")
plt.title("M01_CAS - Post-modifica")
```

### Parametrizzazione e salvataggio su file

```python
plot = M01_CAS_post2
title = "M01_CAS - Post-modifica"
file = "M01_CAS_post.jpg"
sns.scatterplot(data=plot[plot["Valore"]!=0], x="DataOra", y="Valore", 
                size="tentativi", 
                hue="tentativi",
               palette="coolwarm")
plt.title(title)
plt.savefig(file, dpi=300)
```
